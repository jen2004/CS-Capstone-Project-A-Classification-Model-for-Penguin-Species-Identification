```python
# Install necessary libraries
%pip install pandas
%pip install seaborn
%pip install matplotlib

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
from google.colab import files
uploaded = files.upload()
file_path = 'penguin_dataset.csv'
df = pd.read_csv(file_path)

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Create a figure for all histograms
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle('Penguin Dataset Histograms', fontsize=20)

# Plot histogram for species
sns.histplot(df['species'], ax=axes[0, 0], kde=False, color='blue')
axes[0, 0].set_title('Species Distribution')

# Plot histogram for island
sns.histplot(df['island'], ax=axes[0, 1], kde=False, color='green')
axes[0, 1].set_title('Island Distribution')

# Plot histogram for year
sns.histplot(df['year'], ax=axes[0, 2], kde=False, color='purple')
axes[0, 2].set_title('Year Distribution')

# Plot histogram for bill length
sns.histplot(df['bill_length_mm'], ax=axes[1, 0], kde=True, color='orange')
axes[1, 0].set_title('Bill Length Distribution')

# Plot histogram for bill depth
sns.histplot(df['bill_depth_mm'], ax=axes[1, 1], kde=True, color='red')
axes[1, 1].set_title('Bill Depth Distribution')

# Plot histogram for flipper length
sns.histplot(df['flipper_length_mm'], ax=axes[1, 2], kde=True, color='cyan')
axes[1, 2].set_title('Flipper Length Distribution')

# Plot histogram for body mass
sns.histplot(df['body_mass_g'], ax=axes[2, 0], kde=True, color='magenta')
axes[2, 0].set_title('Body Mass Distribution')

# Plot histogram for sex
sns.histplot(df['sex'], ax=axes[2, 1], kde=False, color='brown')
axes[2, 1].set_title('Sex Distribution')

# Hide the last subplot (unused)
axes[2, 2].axis('off')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the title

# Show the plots
plt.show()

# Create a scatterplot matrix (pairplot) comparing all characteristics
# Selecting only numerical columns and 'species' for hue (to differentiate by species)
numerical_columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
df_numerical = df[numerical_columns + ['species']]  # Include 'species' for color grouping

# Create the pairplot
sns.pairplot(df_numerical, hue='species', palette='Set1', markers=["o", "s", "D"], plot_kws={'alpha':0.6})


# Adjust layout to leave space for the pairplot title
plt.subplots_adjust(top=0.95)  # Adjust the top space to prevent overlap
plt.suptitle('Scatterplot Matrix of Penguin Characteristics', fontsize=20)

# Show the scatterplot matrix
plt.suptitle('Scatterplot Matrix of Penguin Characteristics', fontsize=20)
plt.show()

# Create a scatterplot of bill length vs bill depth, separated by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='species', palette='Set1', s=100)

# Add titles and labels
plt.title('Bill Length vs Bill Depth by Species', fontsize=16)
plt.xlabel('Bill Length (mm)', fontsize=12)
plt.ylabel('Bill Depth (mm)', fontsize=12)

# Show the scatterplot
plt.show()


```
